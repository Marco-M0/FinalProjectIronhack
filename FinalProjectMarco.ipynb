```python
# Normalizing values in "Target Value" - step 1, removing % character

def update_target_unit_if_percent(df):
    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        target_value = str(row['Target Value'])  # Convert to string for checking
        
        # Check if the last character is '%'
        if target_value.endswith('%'):
            # Remove the '%' symbol and convert the value back to a number
            new_value = float(target_value.rstrip('%'))
            
            # Update the 'Target Value' without the '%' and the 'Target Unit' with '%'
            df.at[index, 'Target Value'] = new_value
            df.at[index, 'Target Unit'] = '%'
    
    return df

# Apply the function to sbti_df
sbti_df = update_target_unit_if_percent(sbti_df)
```

```python
# Normalizing values in "Target Value" - step 2, removing "°C" characters

def update_target_unit_if_degree_celsius(df):
    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        target_value = str(row['Target Value'])  # Convert to string for checking
        
        # Check if the last two characters are '°C'
        if target_value.endswith('°C'):
            # Remove the '°C' symbol and convert the value back to a number
            new_value = float(target_value.rstrip('°C').strip())  # Remove and strip extra spaces
            
            # Update the 'Target Value' without the '°C' and the 'Target Unit' with '°C'
            df.at[index, 'Target Value'] = new_value
            df.at[index, 'Target Unit'] = '°C'
    
    return df

# Apply the function to sbti_df
sbti_df = update_target_unit_if_degree_celsius(sbti_df)
```

```python
# Normalizing values in "Target Value" - step 3, percentages as decimals
def update_target_value_and_unit(df):
    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        target_value = row['Target Value']
        
        # Check if the value is greater than 0 and less than 1
        if 0 < target_value < 1:
            # Multiply the value by 100
            df.at[index, 'Target Value'] = target_value * 100
            # Update the 'Target Unit' to '%'
            df.at[index, 'Target Unit'] = '%'
    
    return df

# Apply the function to sbti_df
sbti_df = update_target_value_and_unit(sbti_df)
```

```python
sbti_df = sbti_df.fillna('NA')
```

```python
def replace_na_with_zero(df):
    # List of columns to check
    columns_to_replace = ["Commitment Deadline", "Target Value", "Base Year", "Target Year", "Date Published"]
    
    # Replace 'NA' with 0 in the specified columns
    df[columns_to_replace] = df[columns_to_replace].replace('NA', 0)
    
    return df

# Apply the function to sbti_df
sbti_df = replace_na_with_zero(sbti_df)
```

```python
# Convert 'Type Value' column to float
sbti_df['Target Value'] = sbti_df['Target Value'].astype(float)

# Convert 'Base Year' column to int
sbti_df['Base Year'] = sbti_df['Base Year'].astype(int)

# Convert 'Target Year' column to int
sbti_df['Target Year'] = sbti_df['Target Year'].astype(int)

# Convert 'Date Published' column to datetime
sbti_df['Date Published'] = pd.to_datetime(sbti_df['Date Published'], errors='coerce')

# Convert 'Commitment Deadline' column to datetime
sbti_df['Commitment Deadline'] = pd.to_datetime(sbti_df['Commitment Deadline'], errors='coerce')
```

```python
sbti_df['Scope1'] = 0
sbti_df['Scope2'] = 0
sbti_df['Scope3'] = 0
```

```python
def scope_category(df):
    # Iterate through each row in the DataFrame
    for index, row in df.iterrows():
        scope_value = str(row['Scope'])  # Convert the scope value to string
        
        # Check if '1' is in the scope_value
        if '1' in scope_value:
            df.at[index, 'Scope1'] = 1
        
        # Check if '2' is in the scope_value
        if '2' in scope_value:
            df.at[index, 'Scope2'] = 1
        
        # Check if '3' is in the scope_value
        if '3' in scope_value:
            df.at[index, 'Scope3'] = 1
    
    return df

# Apply the function to sbti_df
sbti_df = scope_category(sbti_df)
```

```python
sbti_df['Date Published Year'] = sbti_df['Date Published'].dt.year
sbti_df['Commitment Deadline Year'] = sbti_df['Commitment Deadline'].dt.year
```
